```python
def init():
    return ((3, 3), (0, 0), 'left')
```

```python
def is_reached(state):
    return state == ((0, 0), (3, 3), 'right')
```

```python
def check_valid(current):
    left_bank, right_bank, pos = current

    if 0 <= left_bank[0] <= 3 and 0 <= left_bank[1] <= 3 and 0 <= right_bank[0] <= 3 and 0 <= right_bank[1] <= 3:
        if (left_bank[0] < left_bank[1] and left_bank[0] > 0) or (right_bank[0] < right_bank[1] and right_bank[0] > 0):
            return False
        else:
            return True
    return False
```

```python
def next_states(current):
    left_bank, right_bank, pos = current
    next_states = []

    if pos == 'left':
        left_right_banksible = [(0, 1), (0, 2), (1, 1), (1, 0), (2, 0)]
        for pos in left_right_banksible:
            new_left_bank = (left_bank[0] - pos[0], left_bank[1] - pos[1])
            new_right_bank = (right_bank[0] + pos[0], right_bank[1] + pos[1])
            if check_valid((new_left_bank, new_right_bank, 'right')):
                next_states.append((new_left_bank, new_right_bank, 'right'))
    else:
        right_left_banksible = [(1, 0), (0, 1), (1, 1)]
        for pos in right_left_banksible:
            new_left_bank = (left_bank[0] + pos[0], left_bank[1] + pos[1])
            new_right_bank = (right_bank[0] - pos[0], right_bank[1] - pos[1])
            if check_valid((new_left_bank, new_right_bank, 'left')):
                next_states.append((new_left_bank, new_right_bank, 'left'))

    return next_states
```

```python
def display(algorithm, result):
    if result:
        print('\n' + algorithm + ' Solution:\n')
        for i, state in enumerate(result):
            left_pos, right_pos, boat_pos = state
            print(f"Step {i + 1}: Left Bank: {left_pos} - Right Bank: {right_pos} - Boat Position: {boat_pos}")
    else:
        print("No Solution...")
```

```python
from collections import deque

def bfs(init):
    visited = set()
    queue = deque([(init, [init])])

    while queue:
        current_state, path = queue.popleft()
        print('Current: ', current_state)
        if is_reached(current_state):
            return path

        visited.add(current_state)
        next_possible_states = next_states(current_state)
        print('Next States: ', next_possible_states)
        for next_state in next_possible_states:
            if next_state not in visited:
                queue.append((next_state, path + [next_state]))

    return None

def driver_bfs():
    init_state = init()
    print('\n\nBreadth First Search:\n')
    result = bfs(init_state)
    display('BFS', result)

driver_bfs()
```

```python
def dfs(current_state, path, visited):
    print('Current: ', current_state)
    if is_reached(current_state):
        return path

    visited.add(current_state)
    next_possible_states = next_states(current_state)
    print('Next States: ', next_possible_states)
    for next_state in next_possible_states:
        if next_state not in visited:
            result = dfs(next_state, path + [next_state], visited)
            if result is not None:
                return result

    return None

def driver_dfs():
    init_state = init()
    visited = set()
    path = [init_state]
    print('\n\nDepth First Search:\n')
    result = dfs(init_state, path, visited)
    display('DFS', result)

driver_dfs()
```

```python
def heuristic(state):
    left_bank, right_bank, pos = state
    # We choose the heuristic as the sum of the number of missionaries and cannibals in the right bank
    return right_bank[0] + right_bank[1]
```

```python
from queue import PriorityQueue

def gbfs(init):
    priority_queue = PriorityQueue()
    priority_queue.put((heuristic(init), init))
    visited = set()
    path = []

    while priority_queue:
        current_state = priority_queue.get()[1]
        path = path + [current_state]
        print('Current: ', current_state)

        if is_reached(current_state):
            return path

        if current_state in visited:
            continue

        visited.add(current_state)

        print('Next States: ', next_states(current_state))
        for successor in next_states(current_state):
            if successor not in visited:
                priority_queue.put((heuristic(successor), successor))

    return None

def driver_gbfs():
    init_state = init()
    print('\n\nGreedy Best First Search:\n')
    result = gbfs(init_state)
    display('GBFS', result)

driver_gbfs()
```

```python
def a_star_search(init):
    priority_queue = PriorityQueue()
    priority_queue.put((heuristic(init), init))
    visited = set()
    path = []
    g_values = {init: 0}

    while priority_queue:
        current_state = priority_queue.get()[1]
        path = path + [current_state]
        print('Current: ', current_state)

        if is_reached(current_state):
            return path

        if current_state in visited:
            continue

        visited.add(current_state)

        print('Next States: ', next_states(current_state))
        for next_state in next_states(current_state):
            g_value = g_values[current_state] + 1
            if next_state not in g_values or g_value < g_values[next_state]:
                g_values[next_state] = g_value
                f_value = g_value + heuristic(next_state)
                priority_queue.put((f_value, next_state))

    return None

def driver_a_star_search():
    init_state = init()
    print('\n\nA Star Search:\n')
    result = a_star_search(init_state)
    display('A*', result)

driver_a_star_search()
```
